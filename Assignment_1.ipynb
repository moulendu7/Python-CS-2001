1. Write a program to find the first non-repeating character in a string using a for loop and else.


```python
while True:
    s = input("Enter a string: ").strip()
    if s == "":
        print("Input cannot be empty.\n")
    else:
        break
for ch in s:
    if s.count(ch) == 1:
        print("First non-repeating character:", ch)
        break
else:
    print("No non-repeating character found.")
```

2. Given a number n, write a program to check if it is prime using a for-else loop.


```python
n = input("Enter a number: ").strip()
if not n.isdigit():
    print("Invalid input. Enter a positive integer.")
else:
    n = int(n)

    if n == 1:
        print("1 is not a prime number.")
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                print(n, "is not prime.")
                break
        else:
            print(n, "is prime.")
```

3. Implement a program to find the sum of digits of a number using a while loop and continue statement to skip zeros.


```python
n = input("Enter a number: ").strip()

# basic input validation (no while loop)
if not n.isdigit():
    print("Invalid input. Enter a positive integer.")
else:
    num = int(n)
    total = 0

    while num > 0:
        digit = num % 10
        num //= 10

        if digit == 0:
            continue   # skip zeros

        total += digit

    print("Sum of digits (excluding zeros):", total)
```

5. Given an array of integers, write a program to check if the array contains any duplicate values. Use for with break.


```python
arr_input = input("Enter integers separated by space: ").strip()

# basic validation
if arr_input == "":
    print("Invalid input.")
else:
    arr = list(map(int, arr_input.split()))
    seen = set()

    for num in arr:
        if num in seen:
            print("Array contains duplicates.")
            break
        seen.add(num)
    else:
        print("No duplicates found.")
```

6. Write a program to find the factorial of a number using for loop and demonstrate break when the number is 0.


```python
n = input("Enter a number: ").strip()

# basic input validation
if not n.isdigit():
    print("Invalid input. Enter a positive integer.")
else:
    n = int(n)

    if n == 0:
        print("Factorial of 0 is 1.")
    else:
        fact = 1

        for i in range(1, n + 1):
            if n == 0:
                break
            fact *= i

        print("Factorial of", n, "is", fact)
```

7. Implement a program to search for an element in a list using for-else. If found, print its index; otherwise, print "Not Found".


```python
arr_input = input("Enter list elements (space-separated): ").strip()
target_input = input("Enter element to search: ").strip()

# basic validation
if arr_input == "" or target_input == "":
    print("Invalid input.")
else:
    arr = arr_input.split()
    target = target_input

    for i in range(len(arr)):
        if arr[i] == target:
            print("Found at index:", i)
            break
    else:
        print("Not Found")
```

10. Implement a program that checks if a string is a palindrome using for-else.


```python
s = input("Enter a string: ").strip()

if s == "":
    print("Invalid input.")
else:
    n = len(s)

    for i in range(n // 2):
        if s[i] != s[n - 1 - i]:
            print("Not a palindrome")
            break
    else:
        print("Palindrome")
```

11. Write a program to count vowels in a string using for loop. Use continue to skip consonants.


```python
s = input("Enter a string: ").strip()

if s == "":
    print("Invalid input.")
else:
    vowels = "aeiouAEIOU"
    count = 0

    for ch in s:
        if ch not in vowels:
            continue
        count += 1

    print("Number of vowels:", count)
```

15. Implement a program to reverse a string using for loop.


```python
s = input("Enter a string: ").strip()

if s == "":
    print("Invalid input.")
else:
    rev = ""
    for ch in s:
        rev = ch + rev
    print("Reversed string:", rev)
```

17. Write a program to find the maximum element in a list using for loop.


```python
arr_input = input("Enter numbers (space-separated): ").strip()

if arr_input == "":
    print("Invalid input.")
else:
    arr = list(map(int, arr_input.split()))
    maximum = arr[0]

    for num in arr:
        if num > maximum:
            maximum = num

    print("Maximum element:", maximum)
```

20. Write a program to check if a number is an Armstrong number using for-else.

```python
n = input("Enter a number: ").strip()

if not n.isdigit():
    print("Invalid input.")
else:
    num = int(n)
    power = len(n)
    total = 0

    for ch in n:
        total += int(ch) ** power
    else:
        if total == num:
            print(num, "is an Armstrong number.")
        else:
            print(num, "is not an Armstrong number.")
```
