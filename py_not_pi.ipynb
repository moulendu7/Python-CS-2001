#5 applications of XOR
1.Bitwise Swapping (without a temporary variable):

XOR can swap two numbers without using extra space.

2.Finding the Odd Occurring Element in an Array:

If every element appears twice except one, XOR of all elements gives the unique one (because x ^ x = 0 and x ^ 0 = x).

3.Checksums / Error Detection:

Used in simple error detection (parity bits, LRC, CRC) because XOR highlights bit changes.

4.Encryption / Decryption:

XOR with a key is used in many stream ciphers and lightweight encryption (since cipher = data ^ key and data = cipher ^ key).

5.Detecting Bit Differences:

XOR shows which bits differ between two numbers (1 where bits differ, 0 where same). Useful in Hamming distance and image processing.

implement a function global median that maintains a global list of numbers and return the current median

```python
num =[]
def global_median(n):
  global num
  num.append(n)

  n = len(num)
  mid = n//2
  if n%2==0:
    return (num[mid-1]+num[mid])/2
  else :
    return num[mid]
print(global_median(2))
```
