**Q2**. Write a function bmi(weight, height) that returns the Body Mass Index (BMI).

Formula: BMI = weight / height**2

Example: print(bmi(68, 1.75))

```python

# Function to calculate BMI
def bmi(weight, height):
    return weight / height ** 2

# Validate if the string is a positive float number
def is_positive_number(value):
    # Allow only one decimal point and the rest digits
    if value.count('.') <= 1 and value.replace('.', '', 1).isdigit():
        # Convert to float and check if positive
        return float(value) > 0
    return False

# Get weight input with validation
weight_input = input("Enter your weight in kilograms: ")
while not is_positive_number(weight_input):
    print("Invalid input. Please enter a positive number (no letters or negative values).")
    weight_input = input("Enter your weight in kilograms: ")
weight = float(weight_input)

# Get height input with validation
height_input = input("Enter your height in meters: ")
while not is_positive_number(height_input):
    print("Invalid input. Please enter a positive number (no letters or negative values).")
    height_input = input("Enter your height in meters: ")
height = float(height_input)

# Calculate and print BMI
result = bmi(weight, height)
print("Your BMI is:", round(result, 2))
```

Q3. Write a function calories_burned(duration, rate=10) that returns total calories as duration * rate.
Call the function once with one argument and once with both arguments.

```python
# Define a function to calculate calories burned
def calories_burned(duration, rate=10):
    """
    Calculates total calories burned.

    Parameters:
    duration (float): Duration of activity in minutes
    rate (float): Calories burned per minute (default is 10)

    Returns:
    float: Total calories burned
    """
    return duration * rate

# Custom function to check if input is a valid positive number
def is_positive_number(value):
    # Allow only one decimal point and the rest digits
    if value.count('.') <= 1 and value.replace('.', '', 1).isdigit():
        return float(value) > 0
    return False

# Get valid input for duration
duration_input = input("Enter duration of activity in minutes: ")
while not is_positive_number(duration_input):
    print("Invalid input. Please enter a positive number.")
    duration_input = input("Enter duration of activity in minutes: ")
duration = float(duration_input)  # Convert to float after validation

# Get valid input for custom rate
rate_input = input("Enter calories burned per minute (or press Enter to use default rate of 10): ")

# Check if the user entered a custom rate or left it blank
if rate_input.strip() == "":
    # Call function using only duration (uses default rate)
    print("Total calories burned (default rate):", calories_burned(duration))
else:
    # Validate custom rate input
    while not is_positive_number(rate_input):
        print("Invalid input. Please enter a positive number.")
        rate_input = input("Enter calories burned per minute: ")
    rate = float(rate_input)

    # Call function using both duration and custom rate
    print("Total calories burned (custom rate):", calories_burned(duration, rate))
```

**Q4**. Create a function heart_rate_stats(max_bpm, avg_bpm, resting_bpm) that returns:

‚óè Heart Rate Reserve = max_bpm - resting_bpm

‚óè Average Difference = avg_bpm - resting_bpm

Print both returned values separately.


```python
# Function to calculate Heart Rate Reserve
def calculate_heart_rate_reserve(max_bpm, resting_bpm):
    return max_bpm - resting_bpm


# Function to calculate Average Difference
def calculate_average_difference(avg_bpm, resting_bpm):
    return avg_bpm - resting_bpm


# Function to calculate both stats together
def heart_rate_stats(max_bpm, avg_bpm, resting_bpm):
    heart_rate_reserve = calculate_heart_rate_reserve(max_bpm, resting_bpm)
    avg_difference = calculate_average_difference(avg_bpm, resting_bpm)
    return heart_rate_reserve, avg_difference


# Function to print results
def display_stats(max_bpm, avg_bpm, resting_bpm):
    reserve, difference = heart_rate_stats(max_bpm, avg_bpm, resting_bpm)
    print("\n--- Heart Rate Statistics ---")
    print("Heart Rate Reserve:", reserve)
    print("Average Difference:", difference)


# Function to take validated positive integer input
def get_valid_bpm(prompt):
    while True:
        value = input(prompt)
        if value.isdigit():
            value = int(value)
            if value > 0:
                return value
            else:
                print("Please enter a positive number.")
        else:
            print("Invalid input! Enter only numbers.")


# Take inputs directly (no main function)
print("Enter your Heart Rate values:")
max_bpm = get_valid_bpm("Maximum BPM: ")
avg_bpm = get_valid_bpm("Average BPM: ")
resting_bpm = get_valid_bpm("Resting BPM: ")

# Logical check
if not (resting_bpm < avg_bpm < max_bpm):
    print("\nWarning: Usually, Resting BPM < Average BPM < Maximum BPM. Check your values!")

# Display results
display_stats(max_bpm, avg_bpm, resting_bpm)
```

**Q6**. Create two functions:

‚óè add_calories(c1, c2) that returns sum of calories.

‚óè weekly_burn() that calls add_calories() 7 times for each day‚Äôs burn.
Print total burn.

```python
 # Function to add calories from two workouts
def add_calories(c1, c2):
    return c1 + c2   # Returns the sum of morning and evening calories

# Function to calculate total calories burned in a week
def weekly_burn():
    total = 0   # Initialize total calories to 0
    # Loop runs for 7 days (1 week)
    for i in range(7):
        print(f"\nDay {i+1}:")   # Display the current day number
        # Take calorie inputs for morning and evening workouts
        c1 = float(input("Enter calories from morning workout: "))
        c2 = float(input("Enter calories from evening workout: "))
        # Add daily calories using add_calories() function
        total += add_calories(c1, c2)

    # Print the total calories burned in the entire week
    print("\nTotal weekly calories burned:", round(total,2))

# Call the main function
weekly_burn()
```

**Q11**. Write a function calculate_HRR(max_bpm, resting_bpm, intensity=0.7) that returns Heart Rate
Reserve zone.

```python
# Function to calculate Heart Rate Reserve (HRR) zone
def calculate_HRR(max_bpm, resting_bpm, intensity=0.7):
    hrr = (max_bpm - resting_bpm) * intensity + resting_bpm
    return round(hrr, 2)   # Rounded to 2 decimal places for clarity

maxbpm = input('Enter the maximum BPM : ')
resting = input('Enter the resting BPM : ')

if maxbpm.isdigit() and resting.isdigit():
    maxbpm = int(maxbpm)
    resting = int(resting)
    if maxbpm > resting > 0:
        print("Heart Rate Reserve Zone:", calculate_HRR(maxbpm, resting))
    else:
        print("Invalid: Maximum BPM must be greater than resting BPM and both must be positive.")
else:
    print("Invalid input: Please enter numeric values only.")
```

**Q12**. Write a function daily_burns(calories_list) that prints burn for each day and returns total.

```python
import numpy as np

# Function to print daily burns and return total calories burned
def daily_burns(calories_list):
    # Convert the input list to a NumPy array
    calories_array = np.array(calories_list)

    # Initialize day counter
    day = 1

    # Iterate over each element in the NumPy array
    for i in np.nditer(calories_array):
        print(f'Calories burned on day {day}: {i}')  # Print daily burn
        day += 1  # Increment day counter

    # Calculate total calories burned and return as integer
    total = int(calories_array.sum())
    return total


# Take input from user
user_input = input("Enter calories burned each day separated by spaces: ").split()

# Input validation
if len(user_input) == 0:
    print("Invalid input: List cannot be empty.")
elif not all(item.replace('.', '', 1).isdigit() for item in user_input):
    print("Invalid input: Please enter only numeric values.")
else:
    arr = [float(i) for i in user_input]
    total = daily_burns(arr)
    print('Total Calories burned:', total)
```

**Q17**. Write a recursive function sum_n(n) to return sum of first n natural numbers.
Call sum_n(100).

```python
# Recursive function to calculate sum of first n natural numbers
def sum_n(n):
    if n == 1:        # Base case
        return 1
    else:
        return n + sum_n(n - 1)  # Recursive call

# Call the function for first 100 natural numbers
result = sum_n(100)
print("Sum of first 100 natural numbers:", result)
```

**Q18**. Write a recursive function factorial(n) to return n!

Call factorial(6).

```python
# Recursive function to calculate factorial
def factorial(n):
    if n == 0 or n == 1:  # Base case: 0! = 1! = 1
        return 1
    else:
        return n * factorial(n - 1)  # Recursive call

# Call the function for 6!
result = factorial(6)
print("Factorial of 6 is:", result)
```

**Q22**. Write a function to calculate and print average heart rate by gender.

```python
# Function to calculate average heart rate by gender
def avg_heart_rate_by_gender(genders, avgbpm):
    register = {}  # Empty dictionary to store gender-wise data

    # Step 1: Group Avg_BPM values by gender
    for g, bpm in zip(genders, avgbpm):
        if g not in register:
            register[g] = []  # Initialize list for this gender
        register[g].append(bpm)  # Add bpm to that gender‚Äôs list

    # Step 2: Calculate average for each gender
    for g in register:
       register[g]=round(float(np.array(register[g]).mean()),2)

    return register  # Return final dictionary

# Convert columns to lists
genders = df['Gender'].tolist()
avgbpm = df['Avg_BPM'].tolist()

# Print result
print("Average heart rate by gender:", avg_heart_rate_by_gender(genders, avgbpm))
```

**Q23**. Write a function to visualize Calories_Burned vs Session_Duration for each Workout_Type using
Matplotlib.

```python
def graph_caloriesburned_vs_sessionduration(calories, session, workout):
    # Get all unique workout types in the dataset
    workout_types = np.unique(workout)

    # Loop over each workout type to plot separately
    for w in workout_types:
        # Create a mask (Boolean array) where workout equals the current type
        mask = workout == w

        # Plot calories vs session duration for this workout type
        # marker='o' ‚Üí use circles for points
        # linestyle='' ‚Üí no line connecting points (scatter style)
        # label=w ‚Üí add this workout type to the legend
        plt.plot(session[mask], calories[mask], marker='o', linestyle='', label=w)

    # Add title and axis labels
    plt.title('Calories Burned vs Session Duration by Workout Type')
    plt.xlabel('Session Duration (hours)')
    plt.ylabel('Calories Burned')

    # Show legend to distinguish different workout types
    plt.legend()

    # Display the plot
    plt.show()


# Convert dataframe columns to NumPy arrays for plotting
calories = df['Calories_Burned'].to_numpy()
session = df['Session_Duration (hours)'].to_numpy()
workout = df['Workout_Type'].to_numpy()

# Call the function to plot the data
graph_caloriesburned_vs_sessionduration(calories, session, workout)
```

**Q24**. Write a function to compute correlation between any two numeric columns (column names taken
from user input).

```python


def correlation_bw_2(col1, col2):
    """
    Computes the Pearson correlation coefficient between two columns of a DataFrame.

    Parameters:
    col1 : str
        Name of the first column
    col2 : str
        Name of the second column

    Returns:
    float
        Pearson correlation coefficient rounded to 3 decimal places
    """
    # Convert the first column of the DataFrame to a NumPy array
    c1 = df[col1].to_numpy()

    # Convert the second column of the DataFrame to a NumPy array
    c2 = df[col2].to_numpy()

    # Compute the correlation matrix and extract the correlation coefficient
    # [0,1] gives the correlation between col1 and col2
    corr = np.corrcoef(c1, c2)[0,1]

    # Round the correlation coefficient to 3 decimal places for readability
    return round(corr, 3)

# Take input from the user for the column names
col1 = input('Enter name of 1st column: ')
col2 = input('Enter name of 2nd column: ')

# Call the function and print the result
# Note: It's a single correlation coefficient, not a full matrix
print(f'Correlation coefficient between {col1} & {col2} is: {correlation_bw_2(col1, col2)}')
```

**Q25**. Write a function that creates and saves any Matplotlib plot as a PNG file automatically.

```python
import matplotlib.pyplot as plt

def save_plot_as_png(x, y, title='Plot', filename='plot.png'):
    # Check if x and y have the same length
    if len(x) != len(y):
        print("Graph cannot be plotted :(")
        return  # stop the function if lengths don't match

    # Create a new figure
    plt.figure()

    # Plot the data with circles at points and a solid line
    plt.plot(x, y, marker='o', linestyle='-', color='blue')

    # Add title and axis labels
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Add a grid
    plt.grid(True)

    # Save the plot as a PNG file
    plt.savefig(filename, format='png')

    # Display the plot
    plt.show()
    print() # For a gap

    print(f"Plot saved as '{filename}'")

x = input("Enter elements for X-axis :").split()
y = input("Enter elements for Y-axis :").split()
title = input("Enter a title for the plot :")
filename = input('Enter the filename you want to save :')

save_plot_as_png(x,y,title,filename)


```
